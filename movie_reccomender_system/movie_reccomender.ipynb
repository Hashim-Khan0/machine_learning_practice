```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
movie_titles_df = pd.read_csv('Movie_Id_Titles')
```

```python
movie_titles_df.head()
```

```python
movie_ratings_df = pd.read_csv('u.data', sep = '\t', names = ['user_id', 'item_id', 'rating', 'timestamp'])
```

```python
movie_ratings_df.head()
```

```python
movie_ratings_df.drop(['timestamp'], axis = 1, inplace = True)
movie_ratings_df
```

```python
movie_ratings_df.info()
```

```python
movies_df = pd.merge(movie_ratings_df, movie_titles_df, on = 'item_id')
movies_df
```

```python
movies_df.shape
```

```python
movies_df.groupby('title')['rating'].describe()
```

```python
ratings_df_mean = movies_df.groupby('title')['rating'].describe()['mean']
ratings_df_mean
```

```python
ratings_df_count = movies_df.groupby('title')['rating'].describe()['count']
ratings_df_count
```

```python
ratings_mean_count_df = pd.concat([ratings_df_count, ratings_df_mean], axis = 1)
ratings_mean_count_df.reset_index()
```

```python
ratings_mean_count_df['mean'].plot(bins = 100, kind = 'hist', color = 'orange')
```

```python
ratings_mean_count_df['count'].plot(bins = 100, kind = 'hist', color = 'orange')
```

```python
ratings_mean_count_df[ ratings_mean_count_df['mean'] == 5]
```

```python
ratings_mean_count_df.sort_values('count', ascending = False).head()
```

```python
movies_df
```

```python
userid_movietitle_matrix = movies_df.pivot_table(index = 'user_id', columns = 'title', values = 'rating')
userid_movietitle_matrix
```

```python
titanic = userid_movietitle_matrix['Titanic (1997)']
titanic
```

```python
starwars = userid_movietitle_matrix['Star Wars (1977)']
starwars
```

```python
titanic_correlations = pd.DataFrame(userid_movietitle_matrix.corrwith(titanic), columns = ['Correlation'])
```

```python
titanic_correlations = titanic_correlations.join(ratings_mean_count_df['count'])
```

```python
titanic_correlations.dropna(inplace=True)
```

```python
titanic_correlations.sort_values('Correlation', ascending = False)
```

```python
titanic_correlations[titanic_correlations['count']>80].sort_values('Correlation', ascending=False)
```

```python
userid_movietitle_matrix
```

```python
movie_correlations = userid_movietitle_matrix.corr(method = 'pearson', min_periods = 80)
movie_correlations
```

```python
myRatings = pd.read_csv('My_Ratings.csv')
myRatings
```

```python
similar_movies_list = pd.Series()

for i in range(0,2):
    similar_movie = movie_correlations[  myRatings['Movie Name'][i] ].dropna()
    similar_movie = similar_movie.map(lambda x: x * myRatings ['Ratings'][i])
    similar_movies_list = pd.concat([similar_movies_list, similar_movie])
```

```python
similar_movies_list.sort_values(inplace = True, ascending = False)
print(similar_movies_list.head(10))
```
